Essayons de faire une PCA pour voir ce qu'on obtient.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

original_cols = []
encoded_cols = []

for col in ['Nature mutation', 'Type de voie', 'Voie', 'Code postal', 'Code departement', 'Code commune', 'Type local']:
    df[col] = le.fit_transform(df[col].astype(str))
    original_cols.append(pd.DataFrame({f'{col}_original': le.classes_}))
    encoded_cols.append(pd.DataFrame({f'{col}_encoded': range(len(le.classes_))}))

original_df = pd.concat(original_cols, axis=1)
encoded_df = pd.concat(encoded_cols, axis=1)

merged_df = pd.concat([original_df, encoded_df], axis=1)

merged_df
```

Encoding pour chaque colonne de string.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df)

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k, n_init='auto')
    kmeanModel.fit(df_pca)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'rx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('Elbow Method showing optimal k')
plt.show()
```

Nous voyons que 2 semble être le nombre de clusters optimal.

```python
from matplotlib.colors import ListedColormap

kmeans = KMeans(n_clusters=2, n_init='auto')
kmeans.fit(df_pca)
labels = kmeans.labels_
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels, cmap=ListedColormap(['orange', 'red']))
```

```python
df['kmeans_labels'] = labels
if df.loc[df['kmeans_labels'] == 1].shape[0] > df.loc[df['kmeans_labels'] == 0].shape[0]:
    outlier_cluster = 0
else:
    outlier_cluster = 1
```

```python
outlier_values = df[df['kmeans_labels'] == outlier_cluster]
outlier_values_mean = outlier_values[['Valeur fonciere', 'Surface Carrez du 1er lot', 'Surface Carrez totale', 'Nombre de lots', 'Surface reelle bati', 'Nombre pieces principales']].mean()
```

```python
normal_values = df[df['kmeans_labels'] != outlier_cluster]
normal_values_mean = normal_values[['Valeur fonciere', 'Surface Carrez du 1er lot', 'Surface Carrez totale', 'Nombre de lots', 'Surface reelle bati', 'Nombre pieces principales']].mean()
```

```python
mean_df = pd.DataFrame({'Normal': normal_values_mean, 'Outliers': outlier_values_mean})

fig, axes = plt.subplots(2, 3, figsize=(10, 6)) # Adjust the dimensions to fit the 2x3 grid

for i, feature in enumerate(mean_df.index):
    # Convert the flat index to a grid index
    grid_idx = (i // 3, i % 3)

    axes[grid_idx].bar(['Normal', 'Outliers'], mean_df.loc[feature], alpha=0.5, color=['orange', 'red'])
    axes[grid_idx].set_ylabel('Value')
    axes[grid_idx].set_title(f'Bar Plot of {feature}')

plt.tight_layout()
plt.show()
```

Nous voyons que les moyennes des valeurs sont beaucoup plus grandes pour la valeur foncière, les surfaces Carrez et les surfaces réelles bâties, ce qui n'est pas exactement le cas pour les nombres de lots et les nombres de pièces principales. Nous voyons que les clusters se sont formés principalement en fonction de la valeur foncière lorsqu'on voit les scatters de la surface réelle bâti et/ou la surface Carrez totale en fonction de la valeur foncière. On peut également en tirer cette conclusion juste en observant le scatter de la surface réelle bâti en fonction de la surface Carrez totale, où tous les clusters se retrouvent au même niveau, chacun emmêlé aux 2 autres.

Une explication pour cet incident serait que puisque KMeans fonctionne sur la base de la distance euclidienne, les caractéristiques avec des échelles plus grandes (ce qui est le cas de la valeur foncière) auront une influence disproportionnée sur les résultats.
Nous nous retrouvons avec deux options : soit nous normalisons/standardisons les données, soit nous exploitons ces disparités dans les regroupements à notre avantage.

Nous avons décidé d'en tirer parti de ces disparités.

```python
# Dropping the normals
df = df[df['kmeans_labels'] != outlier_cluster]
df = df.drop(columns='kmeans_labels')

df['new_kmeans_labels'] = labels
```

```python
highest_values_cluster = df.loc[df['Valeur fonciere']==df['Valeur fonciere'].max()]['new_kmeans_labels'].unique()[0]
```

```python
df = df[df['new_kmeans_labels']==highest_values_cluster]
df
```

```python
groupby_df = df.groupby('Code departement')['Valeur fonciere'].count()
groupby_df
```

```python
# Get unique 'Code departement_encoded' for  indexing
unique_merged_df = merged_df.drop_duplicates(subset='Code departement_encoded', keep='first')
# Get the before LabelEncoding Code departement
original_labels = unique_merged_df.set_index('Code departement_encoded')['Code departement_original'].to_dict()
# Revert the encoded labels back to original
groupby_df.index = groupby_df.index.map(original_labels)
```

```python
groupby_df
```

```python
top_3 = groupby_df.nlargest(3)
```

```python
colors = ['red' if val in top_3.values else 'lightblue' for val in groupby_df.values]

groupby_df.plot(kind='bar', figsize=(10,6), color=colors)
plt.xlabel('Code departement')
plt.ylabel('Valeur fonciere')
plt.title('Valeur fonciere par Code departement')
plt.show()
```

Et donc, ici, vous voyez l'histogramme où le ratio Surface Carrez ou Surface réelle bâti sur la valeur foncière est minime. En termes simples, vous voyez les départements où les prix des habitations sont élevées sans forcément qu'il y ait un grand espace pour y vivre.
