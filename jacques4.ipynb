```python
# Drop columns where there's too much missing values
df = df.drop(columns=['Identifiant de document', 'Reference document', '1 Articles CGI', '2 Articles CGI', '3 Articles CGI', '3 Articles CGI', '4 Articles CGI', '5 Articles CGI', 'B/T/Q', 'Prefixe de section', 'No Volume', '2eme lot', '3eme lot', '4eme lot', '5eme lot', 'Surface Carrez du 2eme lot', 'Surface Carrez du 3eme lot', 'Surface Carrez du 4eme lot', 'Surface Carrez du 5eme lot', 'Identifiant local', 'Nature culture speciale', 'Type local'], axis=1)
df
```

```python
# Drop samples where there's missing target
df = df.dropna(subset=['Valeur fonciere'])
df.shape
```

```python
# Drop samples where there's missing 'Surface reelle bati'
df = df.dropna(subset=['Surface reelle bati'])
df.shape
```

```python
# Change 'Valeur fonciere' to float
df['Valeur fonciere'] = df['Valeur fonciere'].str.replace(',', '.')
df['Valeur fonciere'] = df['Valeur fonciere'].astype(float)
df
```

```python
import seaborn as sns

# Consider 'Valeur fonciere' column for outlier removal

# Boxplot before removing outliers
sns.boxplot(x=df['Valeur fonciere'])
plt.title('Boxplot before outlier removal')
plt.show()

# Calculate IQR
Q1 = df['Valeur fonciere'].quantile(0.25)
Q3 = df['Valeur fonciere'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['Valeur fonciere'] >= lower_bound) & (df['Valeur fonciere'] <= upper_bound)]

# Boxplot after removing outliers
sns.boxplot(x=df['Valeur fonciere'])
plt.title('Boxplot after outlier removal')
plt.show()

df.shape
```

```python
missing_values = df.isnull().sum()

plt.figure(figsize=(12, 6))
missing_values.plot(kind='bar')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.title('Missing Values by Column')
plt.show()
```

```python
# Fill missing values in 'Type de voie' by 'Non renseignÃ©'
df['Type de voie'].fillna('Non renseignÃ©', inplace=True)
```

```python
# Groupby department, each department has a dataframe
#'Code type local' == 1:maison, 2:appartement, 3:dependance, 4:local

dfs = {}
# ItÃ©rez sur les groupes crÃ©Ã©s par groupby
for code, group in df.groupby('Code type local'):
    # CrÃ©ez un nouveau dataframe pour chaque groupe et ajoutez-le au dictionnaire
    dfs[code] = group.copy()
```

```python
# Change the keys from code to string/label
key_mapping = {
    1.0: 'maison',
    2.0: 'appartement',
    3.0: 'dependance',
    4.0: 'local'
}
dfs = {key_mapping[key]: value for key, value in dfs.items()}
```

# MAISON

```python
df_maison = dfs['maison']
```

```python
missing_values = df_maison.isnull().sum()

plt.figure(figsize=(12, 6))
missing_values.plot(kind='bar')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.title('Missing Values by Column')
plt.show()
```

```python
# Eliminating lots as there aren't lot in houses
df_maison = df_maison.drop(columns=['1er lot', 'Surface Carrez du 1er lot'])
df_maison
```

```python
df_maison['Type de voie'].fillna('Non renseignÃ©', inplace=True)
df_maison = df_maison.dropna()
df_maison
```

```python
# Encode categorical values to numerical
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

columns_to_encode = ['Date mutation', 'Nature mutation', 'Type de voie', 'Code voie', 'Voie', 'Commune', 'Code departement', 'Section', 'Nature culture']

for column in columns_to_encode:
    df_maison[column] = le.fit_transform(df_maison[column])

df_maison
```

```python
import seaborn as sns

# Compute correlation matrix
corr_matrix = df_maison.corr()

# Plot correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

```python
maison_by_departement= {}

# Iterate over unique values in 'Code departement'
for code_dept, data in df_maison.groupby('Code departement'):
    maison_by_departement[code_dept] = data.copy()

for dep, df_maison_dep in maison_by_departement.items():
    df_maison_dep.fillna('0', inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import spearmanr
from sklearn.preprocessing import StandardScaler

maison_score_par_dept = {}

for dep, df_dep in maison_by_departement.items():

    scaler = StandardScaler()
    df_dep = pd.DataFrame(scaler.fit_transform(df_dep), columns=df_dep.columns)

    # Modeling
    # Define your features and target variable
    X = df_dep.drop(columns='Valeur fonciere')
    y = df_dep['Valeur fonciere']
    # Split the dataset into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

    # Create a Linear Regression model
    model = LinearRegression()
    # Train the model using the training sets
    model.fit(X_train, y_train)
    # Make predictions using the testing set
    y_pred = model.predict(X_test)

    resultats = []

    # Print the mean squared error of the model
    #print('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))
    resultats.append('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))

    # Print the coefficient of determination R^2 of the prediction.
    #print('Coefficient of determination R^2: %.2f' % r2_score(y_test, y_pred))
    resultats.append('Coefficient of determination R^2: %.2f' % r2_score(y_test, y_pred))

    spearman_correlation, _ = spearmanr(y_test, y_pred)
    #print('Spearmans correlation: %.3f' % spearman_correlation)
    resultats.append('Spearmans correlation: %.3f' % spearman_correlation)

    #print('Linear Regression score: ', model.score(X_test, y_test))
    resultats.append('Linear Regression score: {}'.format(model.score(X_test, y_test)))

    maison_score_par_dept[dep] = resultats

    """
    plt.scatter(y_test, y_pred)
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs. Predicted Values')
    plt.show()
    """

maison_score_par_dept
```

```python
import re

d = maison_score_par_dept
# Extract R^2 scores and sort keys
score_key_pairs = []

for key, value_list in d.items():
    # Find the string containing the R^2 score
    for s in value_list:
        if 'Coefficient of determination R^2' in s:
            # Extract the score using regex
            match = re.search(r'Coefficient of determination R\^2: ([\d\.]+)', s)
            if match is not None:
                score = float(match.group(1))
                score_key_pairs.append((score, key))
            else:
                print(f"No match found in string {s}")

# Sort keys by score
sorted_keys = [key for score, key in sorted(score_key_pairs, reverse=True)]

print(sorted_keys)
```
